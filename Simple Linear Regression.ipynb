```python
import pandas as pd
dataf = pd.read_csv("E:bmi.csv")
dataf.info()
```

```python
dataf.describe()
```

```python
dataf = dataf[['Height','Weight']]
dataf
```

```python
X = dataf['Height']
y = dataf['Weight']
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
plt.xlabel("Height")
plt.ylabel("Weight")
```

```python
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size = 0.4, random_state = 30)
```

```python
X_train
```

```python
#coverting 1D to 2D array
import numpy as np
X_train = np.array(X_train).reshape(-1,1)
X_train
X_test = np.array(X_test).reshape(-1,1)
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# taining the model
rs = LinearRegression()
rs.fit(X_train,y_train)
```

```python
#y intercept (y = mx +c)
c = rs.intercept_
c
```

```python
# coefficient

m= rs.coef_
m
```

```python
# making prediction

prediction = m*X_train+c
prediction.flatten()
```

```python
pred_test = rs.predict(X_test)
pred_test
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test,pred_test, color='red')
```
